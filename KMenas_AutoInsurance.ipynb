# Deviding data into categorical and numerical form

```python
# Getting categorica values form the dataset - 

categorical_data = []

for i in df.columns:
    if df.columns.dtype == "O":    
        categorical_data.append(i)

categorical_data    
```

```python
# Getting the numerical values form the dataset -

numerical_data = [feature for feature in df.columns if df[feature].dtype != 'O']
numerical_data
```

```python
# Getting the information about the distribution of the data in dataset

for col in categorical_data:
    print(df[col].value_counts())
    print("-"*40)
```

```python
# Getting the information about the distributional proportion of the data in dataset in percentage

for col in categorical_data:
    print(df[col].value_counts(normalize = True)*100)
    print("-"*50)
```

# Univariate Analysis - 

```python
# plotting the kde plot of each feature of numeric data for Univariate Analysis
# checking for the distribution, sckewness and outliers in the data

plt.figure(figsize=(20,17))
plt.suptitle('Univariate AutoInsurance ', fontsize = 10)

for i in range(0, len(numerical_data)):
    plt.subplot(5,3,i+1)
    sns.kdeplot(x = df[numerical_data[i]], shade = True, color = 'r')
    plt.xlabel(numerical_data[i])
    plt.tight_layout()
```

```python
# Univariate Analysis for categorical data

plt.figure(figsize = (20,15))
plt.suptitle('Categorical Univariate Analysis')

for i in range(0, len(categorical_data)):
    plt.subplot(5,3,i+1)
    sns.countplot(x = df[categorical_data[i]])
    plt.xlabel(categorical_data[i])
    plt.xlim(rotation = 45)
    plt.tight_layout()
    
```

# KMeans clustering for the dataset

```python
from sklearn.cluster import KMeans
```

```python
df
```

```python
df.info()
```

```python
# Normalizing the data
# Function for the normalization of the data

def norm_funt(i):
    x = (i-i.min())/(i.max()-i.min())
    return x
```

```python
# Normalizing the numerical features from the dataset

norm_data = norm_funt(df[numerical_data])
norm_data
```

```python
# finding the best number of cluster for making clusters 

ESS = []
k = list(range(1,10))

for i in k:
    model = KMeans(n_clusters=i)
    model.fit(norm_data)
    ESS.append(model.inertia_)
    
ESS
```

```python
# plotting elbow - curve graph for finding number of clusters value 

plt.plot(k, ESS, 'ro-')
plt.show()
```

```python
# taking number of clusters  = 3

model = KMeans(n_clusters=3)
model.fit(norm_data)
model.labels_
```

```python
# adding the clusters values column to the dataset

df['clust'] = model.labels_
df
```
